```python
def f(x):
    return ((x**2)*np.sin(1/x))


fig, ax = plt.subplots(3,3, figsize=(100, 50))

i = 1
dx = 1e-05
for axi in ax:
    for axii in axi:
        x = np.arange(-.5/i, .5/i, dx)
        y = f(x)
        y2 = np.sin(x)
        axii.plot(x, y)
        axii.set_title(f"[{-.5/i:.3f}, {.5/i:.3f}], dx = {dx}", fontsize=50)
        i += 1

plt.show()
```
